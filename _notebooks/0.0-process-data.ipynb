```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/pingpong-caba.csv")
df.head(3)
```

```python
def create_google_maps_url(row):
    lat = row["lat"]
    long = row["long"]
    return f"https://www.google.com/maps/?q={lat},{long}"


df["google_maps"] = df.apply(create_google_maps_url, axis=1)
df.head(3)
```

```python
df.to_csv("../data/pingpong-caba.csv", index=False)
```
